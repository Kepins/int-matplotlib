```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
#plt.style.use('ggplot')
```

# Read CSV

```python
df = pd.read_csv("data.csv", index_col=0, sep=",")
```

# Bar plot
bar plot: x-number of publications, y-town

```python
gr = df.groupby("town").size().to_frame(name="count")
gr = gr[gr["count"]>500].sort_values("count", ascending=False)
```

```python
# plot
plt.yscale("log")

plt.xticks(rotation=90)

plt.xlabel("Town")
plt.ylabel("Publications")
plt.title("Number of publications")

plt.bar(gr.index, gr["count"])
plt.show()
```

# Scatter plot

```python
gr = df.groupby(["town", "lat", "lon"]).size().to_frame(name="count")
```

```python
plt.xlabel("Longitue")
plt.ylabel("Latitude")
plt.title("Publications")

plt.scatter(gr.index.get_level_values("lon"), gr.index.get_level_values("lat"), s=(gr["count"])/50, c=np.random.rand(len(gr),3))

plt.show()
```

# Histogram
x-number of publications, y-year of publication

```python
years = df[df["year"]< 2023]["year"].to_numpy()
```

```python
plt.yscale("log")

plt.hist(years, bins=25)

plt.xlabel("Year")
plt.ylabel("Publications")
plt.title("Number of publications")

plt.show()
```
